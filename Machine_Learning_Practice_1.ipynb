```python
# 데이터 확인
subway = pd.read_csv("Seoul_Subway.csv", encoding = "euc-kr")
subway
```

```python
subway.describe()
```

```python
#시간대 별 평균 혼잡도
time_columns = subway.columns[6:] 
average_congestion = subway[time_columns].mean()
plt.figure(figsize=(12, 6))
plt.plot(time_columns, average_congestion, marker='o', linestyle='-', color='b', label="average congestion")
plt.xticks(rotation=45)
plt.xlabel("time")
plt.ylabel("average congestion (%)")
plt.title("Average congestion by hour")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#test, train Separation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

selected_times = subway.columns[6:]

features = ["호선", "출발역", "요일구분"]
target = subway[selected_times].mean(axis=1)


le_day = LabelEncoder()
subway["요일구분"] = le_day.fit_transform(subway["요일구분"])

le_station = LabelEncoder()
subway["출발역"] = le_station.fit_transform(subway["출발역"])

X = subway[features]
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1001)

print("Train set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
# Linear Regression fit
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# coef_ and intercept_
print('coef_', lr.coef_)
print('intercept_', lr.intercept_)
```

```python
# R2 score
print('R2 score for train set: {:.3f}'.format(lr.score(X_train, y_train)))
print('R2 score for test set: {:.3f}'.format(lr.score(X_test, y_test)))
```

```python
#predict
predicted = lr.predict(X_test)
fig, ax = plt.subplots()
ax.scatter(y_test, predicted, edgecolors=(0, 0, 0)) 
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
